This notebook takes the gpml files of the rift database associated with Merdith et al. 2019, GPC
and extracts metadata from them.  This was done to get start/end lat/lon points of each rift for the Table in the supp material of Merdith et al. 2019.

References

Merdith, A.S., Williams, S.E., Brune, S., Collins, A.S. and Müller, R.D. 2019. Rift and Plate Boundary Evolution across Two Supercontinent Cycles. Global and Planetary Change, 173, pp.1–14 

```python
import pygplates
```

```python
#set path to rotation file, shapes etc.
basedir = '/Users/Andrew/Documents/PhD/Scripts/Python_Scripts/pyGPlates_examples/Merdith_2019_GPC/Sample_data/'

#load rift file
rifts = pygplates.FeatureCollection('%sMer19_Neoprot_Rifts.gpml' % basedir)
```

```python
#we want to extract name, plate ID, valid time, lat/lon co-ordinates and length in kms
for index, rift in enumerate(rifts):
    print 'rift', index
    print 'name', rift.get_name()
    print 'plateID', rift.get_reconstruction_plate_id()
    print 'valid time', rift.get_valid_time()
    print 'lat/lons', rift.get_geometry().to_lat_lon_array()
    rift_length = rift.get_geometry().get_arc_length()
    individual_rift_length_kms = rift_length * pygplates.Earth.mean_radius_in_kms
    print 'length in kms', individual_rift_length_kms
    print '---------'
```
